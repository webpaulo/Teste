```python
import pandas as pd
df = pd.read_csv("dataset1.csv")
df.head()
```

```python
#1
df.shape
```

```python
df_clean = df.drop(columns=['Unnamed: 0','ChestPain','Thal'])
```

```python
df_clean.head()
```

```python
# 3
df_clean.isna().sum()
```

```python
df_clean = df_clean.dropna()
df_clean.isna().sum()
```

```python
df_clean.head()
```

```python
''' 4 Siga os passos do fluxograma apresentado em aula. Primeiro, vamos dividir o
dataset em dois conjuntos distintos: treino e teste. O treino contém as observações
que utilizaremos para treinar o modelo e o teste contém as observações que
utilizaremos para verificar o desempenho do modelo. Utilize o parâmetro
random_state=42 ao fazer essa divisão. '''
from sklearn.model_selection import train_test_split

y = df_clean['AHD']
print(y)
```

```python
df_clean.head()
```

```python
#X= df_clean[:-1]
X = df_clean.drop(columns=['AHD'])
print(X)

# using the train test split function
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2, shuffle=True)
```

```python
print('X_train : ')
print(X_train.head())
print('')
print('X_test : ')
print(X_test.head())
print('')
print('y_train : ')
print(y_train.head())
print('')
print('y_test : ')
print(y_test.head())
```

```python
# 5.a
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB().fit(X_train, y_train)

# funcão para score e erro
AcertoIn = clf.score(X_train, y_train)
AcertoOut = clf.score(X_test, y_test)
print(f"Acerto dentro da amostra: {AcertoIn}")
print(f"Acerto fora da amostra: {AcertoOut}")
# O erro é complemento do acerto
Ein = 1 - AcertoIn
Eout = 1 - AcertoOut
print(f"Erro dentro da amostra: {Ein}")
print(f"Erro fora da amostra: {Eout}")
```
